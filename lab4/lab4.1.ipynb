```python
import pandas as pd

# Завантаження даних та обробка порожніх значень
data = pd.read_csv('household_power_consumption.txt', delimiter=';', na_values='?', header=0)
data.dropna(inplace=True)

# Перетворення стовпців у тип даних float
columns_to_convert = ['Global_active_power', 'Global_reactive_power', 'Voltage', 
                      'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
data[columns_to_convert] = data[columns_to_convert].astype(float)

data['Date'] = data['Date'].astype(str)
data['Time'] = data['Time'].astype(str)

data.head()
```

1. Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
import pandas as pd
import numpy as np
import timeit

start_time_df = timeit.default_timer()
households_df = data[data['Global_active_power'] > 5]  
end_time_df = timeit.default_timer()

print("Час виконання для DataFrame:", end_time_df - start_time_df)
households_df.head()
```

```python
import numpy as np
import timeit

# Заміна NaN у вихідних даних нулями для numpy array
data_np = np.nan_to_num(data.to_numpy())

start_time_np = timeit.default_timer()
households_np = data_np[data_np[:, 2] > 5] 
end_time_np = timeit.default_timer()

print("Час виконання для numpy array:", end_time_np - start_time_np)
print(households_np[:5])
```

2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.

```python
start_time_df = timeit.default_timer()
households_df = data[data['Voltage'] > 235]
end_time_df = timeit.default_timer()

print("Час виконання для DataFrame:", end_time_df - start_time_df)
households_df.head()
```

```python
start_time_np = timeit.default_timer()
data_np = np.array(data)
households_np = data_np[data_np[:, 4] > 235]  
end_time_np = timeit.default_timer()

print("Час виконання для numpy array:", end_time_np - start_time_np)
print(households_np[:5])
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
start_time_df = timeit.default_timer()
households_df = data[(data['Global_intensity'] >= 19) & (data['Global_intensity'] <= 20) 
                     & (data['Sub_metering_2'] > data['Sub_metering_3'])]

end_time_df = timeit.default_timer()

print("Час виконання для DataFrame:", end_time_df - start_time_df)
households_df.head()
```

```python
start_time_np = timeit.default_timer()
data_np = np.array(data)

households_np = data_np[(data_np[:, 5] >= 19) & (data_np[:, 5] <= 20) & 
                              (data_np[:, 7] > data_np[:, 8])]
end_time_np = timeit.default_timer()

print("Час виконання для numpy array:", end_time_np - start_time_np)
print(households_np[:5])
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії.

```python
random_indices = np.random.choice(len(data), size=500000, replace=False)
households_df = data.iloc[random_indices]
start_time_df = timeit.default_timer()
groups_df = households_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
end_time_df = timeit.default_timer()

print("Час виконання для DataFrame:", end_time_df - start_time_df)
groups_df.head()
```

```python
random_indices = np.random.choice(len(data), size=500000, replace=False)
households_np = data.to_numpy()[random_indices]
start_time_np = timeit.default_timer()
groups_np = np.mean(households_np[:, 6:9], axis=0)
end_time_np = timeit.default_timer()

print("Час виконання для numpy array:", end_time_np - start_time_np)
print(groups_np)
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

```python
start_time_df = timeit.default_timer()
# Вибір домогосподарств, які після 18:00 споживають понад 6 кВт за хвилину в середньому
households_df = data[(data['Time'] > '18:00:00') & (data['Global_active_power'] > 6) &
(data['Sub_metering_2'] > data['Sub_metering_1']) & (data['Sub_metering_2'] > data['Sub_metering_3'])]

# Вибір кожного третього результату із першої половини та кожного четвертого результату із другої половини
households_df_1 = households_df.iloc[:len(households_df) // 2:3]
households_df_2 = households_df.iloc[len(households_df) // 2::4]
result_df = pd.concat([households_df_1, households_df_2])

end_time_df = timeit.default_timer()
print("Час виконання для DataFrame:", end_time_df - start_time_df)
result_df.head()


```

```python
import numpy as np
import pandas as pd
import timeit
from datetime import datetime
data['Time'] = pd.to_timedelta(data['Time'])
data_np = data.to_numpy()
start_time_np = timeit.default_timer()
time_threshold = np.timedelta64(18, 'h')

households_np = data_np[(data_np[:, 1] > time_threshold) & 
                              (data_np[:, 2] > 6) & 
                              (data_np[:, 7] > data_np[:, 6]) & 
                              (data_np[:, 7] > data_np[:, 8])]

households_np_1 = households_np[:len(households_np) // 2:3]
households_np_2 = households_np[len(households_np) // 2::4]
result_np = np.concatenate([households_np_1, households_np_2])
end_time_np = timeit.default_timer()
print("Час виконання для numpy array:", end_time_np - start_time_np)
print(result_np)
```
